### Jak napsat string?

Máme několik cest? Obecně mezi nimi neni rozdíl.

````python
print('Petr')
print("Petr")
print('''Petr''')  # víceřádkový zápis - možné ale nedoporučené
print("""Petr""")  # víceřádkový zápis - doporučený zápis
````

````python
print(type('Petr'))
print(type("Petr"))
print(type('''Petr'''))  # víceřádkový zápis - možné ale nedoporučené
print(type("""Petr"""))  # víceřádkový zápis - doporučený zápis
````

Nejčastěji se používá `"` a `"""`.

**Jak to ale vypadat nemá?** Uvozovky jsou nutné (jinak ten zápis pro Python má jiný význam a nejspíše skončí chybou).

````python
print(Petr)
````

Také neni možné uvozovky různě kombinovat. Vždy je nutné, aby byla stejná uvozovka na začátku a na konci řetezce (ohraničuje ho). V řetezci se také nesmí tento druh uvozovky dále vyskytovat (to může být důvod k použití `'` jako ohraničení řetezce pokud chceme v řetezci používat `"`).

````python
print('Petr")
print('Petr")
print(''Petr'')
print(""Petr"")
````

````python
print("I'''''m")
````

````python
print('I"""""m')
````

| \b |	*Backspace* |
| \f |	*Form feed* |

## Proměnná v Pythonu

---

Proměnná je objekt, který slouží ke **uchování hodnoty** pro pozdější použití.

Pokud máš tedy hodnoty, které chceš použít **více než jedenkrát**, **ulož je** do proměnné. Je to jako pojmenování "obalu" ve kterém je datový typ. Když už má jméno tak ho snáze najdeš a můžeš ten "obal" přepoužít.

Pokud chceš hodnotu použít **pouze jednou**, **nemusíš ji ukládat**.

### Standardní zápis proměnné s hodnotou

Pokud chceš interpretovi oznámit, že chceš hodnotu schovat na později, učiníš tak pomocí symbolu `=`:
    

````python
jmeno = "Petr"
vek = 18

print("Petr", 18)
print(jmeno, vek)
````

Označení `jmeno, vek` jsou jen **jména, označení, identifikátory**, které označí hodnoty `"Petr"` a `18` u tebe v paměti.

````python
print(jmeno)
````

### Pravidla pro pojmenování objektů

Jméno proměnné (někdy také odkaz) **může obsahovat** tyto znaky:
1. **Písmenné** znaky,
2. **Číselné** znaky,
3. **Podtržítka**.

````python
jmeno = "Petr"
jmeno2 = "Petr"
jmeno_2 = "Petr"
Jmeno = "Petr"
jmenoUzivatele = "Petr"
jmeno_uzivatele = "Petr"
````

Existují ovšem znaky, které jméno proměnné **obsahovat nesmí**:
1. Jméno proměnné nesmí **začínat číselným znakem**,
2. jméno proměnné nesmí **obsahovat speciální znaky** (kromě podtržítka),
3. jméno proměnné nesmí **obsahovat mezery**.

Obvykle se nedoporučuje začínat jména **odkazů** pomocí **podtržítka** (pokud nevíš proč to dělat, raději to nedělej):

````python
1jmeno = "Petr"
````

````python
n@me = "Petr"
````

````python
na me = "Petr"
````
